1. Use matplotlib to plot the following equation:

y=x**(2)âˆ’x+2
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2 - x + 2
x = np.linspace(-5, 5, 100)
y = f(x)
```

```python

plt.plot(x,y)
plt.annotate('origin', xy=(0, 0), xytext=(0.5, 10))
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y=âˆšx

y=x3

y=2x

y=1/(x+1)

```python
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(-5, 5, 100)
x3 = np.linspace(-2, 2, 100)
x4 = np.linspace(-5, 5, 100)
y1 = np.sqrt(x1)
y2 = x2 ** 3
y3 = 2 ** x3
y4 = 1 / (x4 + 1)
```

```python
plt.title('Square Root of X')
plt.plot(x1,y1)
plt.show()

plt.title('X times 3')
plt.plot(x2,y2)
plt.show()

plt.title('X times 2')
plt.plot(x3,y3)
plt.show()

plt.title('X plus 1 divided by 1')
plt.plot(x4,y4)
plt.plot()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.subplot(141)
plt.title('Square Root of X')
plt.plot(x1,y1)

plt.subplot(142)
plt.title('X times 3')
plt.plot(x2,y2)

plt.subplot(143)
plt.title('X times 2')
plt.plot(x3,y3)

plt.subplot(144)
plt.title('X plus 1 divided by 1')
plt.plot(x4,y4)
plt.plot()
```

4.Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python

plt.plot(x1,y1,c='plum',label='square root of X')


plt.plot(x2,y2,c='cornflowerblue',label='X times 3')


plt.plot(x3,y3,c='salmon',label='X times 2')


plt.plot(x4,y4,c='silver',label='X plus 1 divided by 1')
plt.legend()
```

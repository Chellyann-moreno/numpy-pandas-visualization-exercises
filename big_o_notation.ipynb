```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.special as sc
```

```python
x=np.array([1,10,100])
```

```python
y1=(0*x)+1
y2=np.log(x)
y3=x
y4=x*np.log(x)
y5=x**2
y6=2**x
y7=sc.factorial(x)
y8=x**x
```

```python
plt.figure(figsize=(12,6))
plt.title('Big O Notation')
plt.xlabel('Element')
plt.ylabel('Operations')
plt.plot(x,y1,alpha=1, label="O(1)")
plt.plot(x,y2,alpha=.5,label="O(log n)")
plt.plot(x,y3, label="O(n)")
plt.plot(x,y4, label= "O(n log n)")
plt.plot(x,y5, label="O(n^2)")
plt.plot(x,y6, label= "O(2^n)")
plt.plot(x,y7, label= "O(n!)")
plt.plot(x,y8, label= "O(n^n)")

plt.legend()
plt.show()
```

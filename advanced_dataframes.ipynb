```python
url=get_db_url('employees')
```

```python
employees=pd.read_sql('Select * from employees',url)
titles=pd.read_sql('Select * from titles',url)
```

```python
#5
m_emp_titles=pd.merge(employees,titles, how='inner', on='emp_no')
m_emp_titles
```

```python

m_emp_titles.groupby('title').hire_date.max()
```

```python
max_hire_dates = m_emp_titles.groupby('title').hire_date.max()
max_hire_dates = pd.DataFrame(max_hire_dates).reset_index()
max_hire_dates
```

```python
#6
pd.merge(m_emp_titles, max_hire_dates, on=['title','hire_date']).sort_values('title')
```

```python
chipotle.item_price = chipotle.item_price.str.strip('$').astype(float)
```

```python
chipotle.info()
```

```python
orders_4 = chipotle.groupby('item_name').quantity.sum().nlargest(4)
orders_4
```

```python
prices = chipotle.groupby('item_name').item_price.sum()
prices.head(6)
```

```python
orders_price = pd.merge(orders_4,prices, on='item_name', how='inner')
orders_price = orders_price.reset_index()
orders_price
```

```python
sns.barplot(data=orders_price, y='item_name', x='item_price')
plt.title('Popular Chipotle Food & Drinks')
plt.xlabel('revenue generated')
plt.show()
```
